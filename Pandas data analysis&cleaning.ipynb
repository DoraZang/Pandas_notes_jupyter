#### Filtering data

```python
df['last']=='Doe' # Filtering like this alone returns a boolean Series (mask).
```

```python
filt=(df['last']=='Doe')
df[filt] # First, assign the boolean mask to a variable, then pass it to the DataFrame. The result is a filtered DataFrame.
```

```python
df.loc[filt] # Yields the same result as above, using .loc is generally preferred for boolean indexing.
```

```python
df.loc[filt,'email'] # Based on the filter, select and view only the 'email' column.
```

```python
filt=(df['last']=='Doe') & (df['first']=='John') # Use the '&' operator for 'AND' conditions.
```

```python
df.loc[filt,'email']
```

```python
filt1=(df['last']=='Schafer') | (df['first']=='John')
df.loc[filt1,'email']
```

```python
df.loc[~filt1,'email'] # Use the '~' operator to negate the filter (i.e., return the opposite results).
```

#### Handling Missing Data & Casting

```python
df.dropna() # Drop rows that contain any None or NaN values.
```

```python
df.dropna(axis='index', how='any') # Same effect, these are the default parameter values.
```

```python
df.dropna(axis='index', how='all') # Setting 'how' to 'all' means a row is only dropped if ALL of its values are None or NaN.
```

```python
df.dropna(axis='columns', how='all')  
# Setting 'axis' to 'columns' and 'how' to 'all' means a column is only dropped if ALL of its values are None or NaN.
```

```python
df.dropna(axis='columns', how='any')
```

```python
df.dropna(axis='index', how='any',subset=['email']) 
# Drop the row only if there is a missing value in the 'email' column. Missing values in other columns are ignored.
```

```python
df.dropna(axis='index', how='all',subset=['last','email'])
# Drop the row only if BOTH 'last' and 'email' are missing values. If one is present, the row is kept.
```

```python
df.replace('NA',np.nan,inplace=True)
df.replace('Missing',np.nan,inplace=True)
df
```

```python
df.isna() # Return a boolean mask indicating whether each value is NA (missing).
```

```python
df.fillna('Missing') # Replace all missing values (NaN/None) with the string 'Missing'.
```
